```python
import pandas as pd
import os 
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from matplotlib  import pyplot as plt
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
import time
import random
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
from sklearn.metrics import recall_score
df = pd.read_csv('data.csv')
df.columns = [i.lower().replace(' ','_') for i in df.columns]
pd.options.display.width=None
```

```python
target =  'default_payment_next_month'
X = df.drop(target,axis = 1)
y = df[target]
```

```python
model = GradientBoostingClassifier(random_state=1)
model.fit(X,y)
```

```python
def prediction_frame(X,y,model,thresh):
    def get_predictions():
        predictions = [1 if i>thresh else 0 for i in model.predict_proba(X)[:,1]] 
        return predictions
    output = pd.DataFrame()
    output['predictions'] = get_predictions() 
    output['actual'] = y
    final = pd.concat([X,output],axis = 1)
    return final 
```

```python
pr_frame = prediction_frame(X,y,model,0.6)
pr_frame
```

```python
def prediction_pattern(pr_frame,col, ncuts):
    sorted_ = pr_frame.sort_values(by = [col],ascending= True)
    sorted_['group'] = pd.qcut(sorted_[col],ncuts)
    ratios = []
    groups = sorted_['group'].unique()
    lower_limits = [str(i).split(',')[0][1::] for i in groups]
    upper_limits = [str(i).split(',')[1][::-1][1::][::-1] for i in groups]
    cat_ranges = [f'{int(float(i))}-{int(float(j))}' for i,j in list(zip(lower_limits,upper_limits))]
    for i in sorted_['group'].unique():
        df = sorted_.groupby(by= 'group').get_group(i)
        ratios.append((df['predictions'].sum()/len(df['actual']))*100)

    
    plt.xticks(rotation = '45')
    plt.box(False)
    plt.plot(cat_ranges,ratios,color = 'maroon')
    for index,data in enumerate(ratios):
        plt.text(x=index , y =data+0.09, s= np.round(data,2) , fontdict=dict(fontsize=10))
    plt.show()
```

```python
prediction_pattern(pr_frame = pr_frame,col= 'age',ncuts = 6)
```
